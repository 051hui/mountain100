```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import platform

warnings.filterwarnings('ignore')

# 운영체제별 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin':  # macOS
    plt.rcParams['font.family'] = 'AppleGothic'
else:  # Linux
    plt.rcParams['font.family'] = 'NanumGothic'

# 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
# -----------------------------------------------------
# 히스토그램
# -----------------------------------------------------
def draw_hist(df, 
              col : str, 
              bins : int = 30,
              figsize = (10, 6),
              color = '#7ae07a',):
    '''
    Hist 함수
    '''
    plt.figure(figsize= figsize)
    
    sns.histplot(data=df, 
                 x=col,
                 color=color, 
                 bins = bins,
                 kde=True)
    
    plt.title(f"'{col}' 히스토그램", fontsize=25)
    plt.xlabel(col) # x축 라벨도 변수명으로 변경
    plt.ylabel('빈도 (Frequency)')
    plt.grid(True, alpha=0.3)
    plt.show()

# -----------------------------------------------------
# 바이올린 + 박스플롯
# -----------------------------------------------------
def draw_violin_box_overlay(df, col, viol_color = "#7ae07a", box_color = "#418341"):
    '''
    violin + Box
    '''

    plt.figure(figsize=(10, 6))

    sns.violinplot(data=df, x=col, color=viol_color, inner=None)
    sns.boxplot(data=df, x=col, width=0.05, color=box_color)

    plt.title(f"'{col}' 분포 (Violin + Box)", fontsize=25)
    plt.xlabel(col)
    plt.grid(True, axis='x', alpha=0.3)
    plt.savefig(f"{col}_overlay_plot.png", transparent=True)
    
    plt.show()

```

```python
df = pd.read_csv("df_하이킹_유클리드_6.csv")
```

```python
# -----------------------------------------------------
# 산 특성 Hist
# -----------------------------------------------------

draw_hist(df, '최고고도_m', 30)
draw_hist(df, '누적상승_m', 30)
draw_hist(df, '총거리_km', 30)
draw_hist(df, '예상시간_분', 30)
```

```python
# -----------------------------------------------------
# 점수 Hist
# -----------------------------------------------------

draw_hist(df, '난이도점수', 30)
draw_hist(df, '관광인프라', 30)
```

```python
draw_hist(df, '매력종합점수', 15)
```

```python
draw_violin_box_overlay(df, '주차장_접근성')
draw_violin_box_overlay(df, '정류장_접근성')
```
